```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
import os
```

```python
model = YOLO("runs/detect/train/weights/best.pt")
```

```python
test_images_path = "/Users/ahsennuraldas/Desktop/LicensePlateDetection/valid/images"
image_files = [f for f in os.listdir(test_images_path) if f.endswith((".jpg", ".png", ".jpeg"))]
```

```python
for image_file in image_files:
    image_path = os.path.join(test_images_path, image_file)

    # Tahmin yap
    results = model.predict(image_path, save=False, conf=0.25)

    # GÃ¶rseli oku
    img = cv2.imread(image_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # BBox'larÄ± Ã§iz
    for r in results:
        for box in r.boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            confidence = float(box.conf[0])
            label = int(box.cls[0])
            cv2.rectangle(img, (x1, y1), (x2, y2), (0,255,0), 2)
            cv2.putText(img, f"{model.names[label]} {confidence:.2f}", (x1, y1 - 10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 0, 0), 2)

    # Matplotlib ile gÃ¶ster
    plt.figure(figsize=(6, 6))
    plt.imshow(img)
    plt.title(f"Tahmin: {image_file}")
    plt.axis("off")
    plt.show()
```

```python
metrics = model.val()
```

```python
from PIL import Image

# Confusion matrix gÃ¶rseli yolunu ayarla
cm_path = "runs/detect/val/confusion_matrix.png"

# GÃ¶ster
img = Image.open(cm_path)
plt.figure(figsize=(6,6), dpi=300)
plt.imshow(img)
plt.axis('off')
plt.title("Confusion Matrix")
plt.show()
```

```python
# Precision
precision = metrics.box.mp
# Recall
recall = metrics.box.mr
# mAP@0.5
map50 = metrics.box.map50
# mAP@0.5:0.95
map = metrics.box.map
```

```python
print(f"Precision:      {precision:.4f}")
print(f"Recall:         {recall:.4f}")
print(f"mAP@0.5:        {map50:.4f}")
print(f"mAP@0.5:0.95:   {map:.4f}")
```

```python
import matplotlib.pyplot as plt

labels = ["Precision", "Recall", "mAP@0.5", "mAP@0.5:0.95"]
values = [precision, recall, map50, map]

plt.figure(figsize=(6,4))
plt.bar(labels, values, color=["blue", "green", "orange", "red"])
plt.ylim(0, 1)
plt.title("YOLOv8 Performans Metrikleri")
plt.ylabel("Skor")
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Orijinal resmi tekrar oku (Ã§Ã¼nkÃ¼ .plot() sonrasÄ± Ã§izilmiÅŸ resim oluyor)
from ultralytics import YOLO
import cv2

# Resmi oku
image_path = '/Users/ahsennuraldas/Desktop/images.jpeg'
original_img = cv2.imread(image_path)

# YOLOv8 ile tahmin al (tekrar)
results = model(image_path)

# Ä°lk tespit sonucundaki bounding box'Ä± al (ilk obje varsayÄ±larak)
boxes = results[0].boxes
xyxy = boxes.xyxy[0].cpu().numpy().astype(int)  # [x1, y1, x2, y2]

# ROI (Region of Interest) = plaka bÃ¶lgesi
plate_img = original_img[xyxy[1]:xyxy[3], xyxy[0]:xyxy[2]]

# OCR iÃ§in RGB formatÄ±na Ã§evir
plate_rgb = cv2.cvtColor(plate_img, cv2.COLOR_BGR2RGB)

# GÃ¶rselleÅŸtir (isteÄŸe baÄŸlÄ±)
from matplotlib import pyplot as plt

plt.imshow(plate_rgb)
plt.title("Tespit Edilen Plaka")
plt.axis('off')
plt.show()
```

```python
# OCR ile plakayÄ± oku
text = pytesseract.image_to_string(plate_rgb, config='--psm 8')  # psm 8: tek bir kelime veya plaka iÃ§in ideal
print("Tespit Edilen Plaka:", text.strip())
```
